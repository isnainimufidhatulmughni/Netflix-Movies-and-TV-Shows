```python
# @title Prepare Data text
#Read data dari spreadsheet di gdrive

sheet_url = 'https://docs.google.com/spreadsheets/d/1ScmiqoudXRHvyS45FWoWrMuSo6A6DOo2XLXqGCMaFY0/edit#gid=116324769'
sheet_url_download = sheet_url.replace ('/edit','/export?format=csv')

df_user = pd.read_csv(sheet_url_download)
df_user.head()
```

```python
# @title Data Cleaning text
#Make a copy of the original data for data cleaning
dfc = df_user.copy()
```

```python
# Check table info
dfc.info()
```

We will change the id data type from int into string

```python
dfc['show_id'] = dfc['show_id'].astype(str)
dfc['type'] = dfc['type'].astype(str)
dfc['title'] = dfc['title'].astype(str)
```

```python
dfc['release_year'] = pd.to_datetime(dfc['release_year'])
dfc['date_added'] = pd.to_datetime(dfc['date_added'])
```

```python
dfc['month_added'] = dfc['date_added'].dt.strftime('%B')
dfc['release_year'] = dfc['release_year'].dt.strftime('%Y')
```

```python
dfc['duration'] = dfc['duration'].str.replace('min', '')
```

```python
dfc['year_added'] = dfc['date_added'].dt.strftime('%Y')
```
